```python
import os
import folium
import json
```

```python
california_counties = os.path.join('../data', 'california_counties.json')
geo_json_data = json.load(open(california_counties))
```

```python
# A first way of drawing it on a map, is simply to use folium.GeoJson :
m = folium.Map([38, -120], zoom_start=6)
folium.GeoJson(geo_json_data).add_to(m)
m.save(os.path.join('results', 'GeoJSON_and_choropleth_0.html'))
m
```
